## Ejercicio 1


```r
f_X = function(x){
    ans = 0
    if(-1<= x & x<=1) ans = 0.75*(1-x^2)
    ans
}

F_X = function(x){
    ans = 0 
    if(-1<= x & x<=1){
        ans = 0.75*x-(x^3)/4+0.5
    } 
    if(1<x) {
        ans = 1
    }
    ans
}

x_range = seq(-1, 1, 0.1)

plot(x_range, sapply(x_range, f_X), type = "o", col = "red", main = "f_X - densidad y F_X - acumulada",
     xlab = "R_X", ylim=c(0,1))

lines(x_range, sapply(x_range, F_X), type = "o", col = "blue")

legend(0, 0.1, legend=c("f_X", "F_X"),
       col=c("red", "blue"), lty=1)
```

## Ejercicio 2

```r
f_X = function(x){
    ans = 0
    if(0<= x & x<=2) ans = 0.375*(x^2)
    ans
}

F_X = function(x){
    ans = 0 
    if(0<= x & x<=2){
        ans = (x^3)/8
    } 
    if(2<x) {
        ans = 1
    }
    ans
}

x_range = seq(0, 2, 0.1)

plot(x_range, sapply(x_range, f_X), type = "o", col = "red", main = "f_X - densidad, F_X - acumulada y mediana",
     xlab = "R_X", ylim=c(0,1))

lines(x_range, sapply(x_range, F_X), type = "o", col = "blue")

abline(v = 4^(1/3), col = "green")

legend(0, 0.6, legend=c("f_X", "F_X", "mediana"),
       col=c("red", "blue", "green"), lty=1)
```
