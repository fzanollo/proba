# Ejercicio 1


```r
alfajores <- read.table("alfajores.txt", header = TRUE)
table(alfajores)
```

```r
totalA = length(alfajores$fabrica)

cat("a) prob(Quilmes):")
sum(alfajores$fabrica == 0) / totalA

cat("b) prob(Quilmes y 3 defectuosos):")
sum(alfajores$fabrica == 0 & alfajores$defectuosos == 3) / totalA

cat("c) prob puntual conjunta del vector XY:")
table(alfajores)/totalA

cat("d) esperanza y var de X:")
mean(alfajores$defectuosos)
var(alfajores$defectuosos)

cat("e) prob(3 defectuosos | Quilmes):")
sum(alfajores$fabrica == 0 & alfajores$defectuosos == 3) / sum(alfajores$fabrica == 0)

cat("f) prob(X=3):")
sum(alfajores$defectuosos == 3) / totalA

cat("g) prob(Quilmes | 3 defectuosos):")
sum(alfajores$fabrica == 0 & alfajores$defectuosos == 3) / sum(alfajores$defectuosos == 3)

cat("h) prob(Pilar | 3 defectuosos):")
sum(alfajores$fabrica == 1 & alfajores$defectuosos == 3) / sum(alfajores$defectuosos == 3)
```

# Ejercicio 6


```r
makeQQ = function(FUN, media, ...){
    dat = data.frame(valores=c(FUN(25, ...), FUN(50, ...), FUN(100, ...)), id=c(rep('a_25',25), rep('b_50',50), rep('c_100', 100)))
    
    ggplot(dat, aes(x=id, y=valores, fill=id)) + 
      geom_boxplot() + coord_flip() + geom_hline(yintercept = media, linetype="dashed", color = "red", size=1)
}
```

```r
cat('a)')
makeQQ(rnorm, 0)
makeQQ(rnorm, 0)
```

```r
cat('b)')
makeQQ(rgamma, 5/0.5, shape = 5, rate = 0.5)
```

```r
cat('c)') #Y= Z/U
yRandomGen = function(n){
    rnorm(n)/runif(n)
}
makeQQ(yRandomGen, 0)
```

```r
cat('d)')
makeQQ(runif, 0.5)
```

```r
cat('e)')
makeQQ(rexp, 1)
```

# Ejercicio 7


```r
datos = runif(10)

l2 = function(xs, c){
    sum((xs-c)^2)
}

l1 = function(xs, c){
    sum(abs(xs-c))
}
```

```r
#a) se minimiza para la media
media = mean(datos)

xs = seq(media-2, media+2, by=0.2)

plot(xs, sapply(xs, l2, xs=datos), main='a) se minimiza para la media', ylab='l2')
abline(v=media, col="red")
```

```r
#b) se minimiza para la mediana, te debo la demo formal
mediana = median(datos)

xs = seq(mediana-2, mediana+2, by=0.2)

plot(xs, sapply(xs, l1, xs=datos), main='b) se minimiza para la mediana', ylab='l1')
abline(v=mediana, col="red")
```

# Ejercicio 9


```r
departamentos <- read.csv("departamentos-en-venta-2016.csv", sep=';')
colnames(departamentos)
```

```r
#a) precio usd vs m2 cubiertos
plot(departamentos$M2CUB, departamentos$DOLARES, xlab='m2 cubiertos', ylab='precio usd', main='a)')
```

```r
#b)
cat('b) correlacion entre X=m2 cubiertos Y= precio usd :', cor(departamentos$M2CUB, departamentos$DOLARES))

#c y d) lm: El primer argumento de este comando es una fórmula y ~ x en la que se especifica cuál es la 
# variable respuesta o dependiente (y) y cuál es la variable regresora o independiente (x). [https://rpubs.com/joser/RegresionSimple]
cat('\n\nc y d) recta de regresion Y vs X :')
regresion = lm(DOLARES ~ M2CUB, data = departamentos)
summary(regresion)

plot(departamentos$M2CUB, departamentos$DOLARES, xlab='m2 cubiertos', ylab='precio usd', main='c)')
abline(regresion, col='red')
```
