# Ejercicio 1


```r
alfajores <- read.table("alfajores.txt", header = TRUE)
table(alfajores)
```

```r
totalA = length(alfajores$fabrica)

cat("a) prob(Quilmes):")
sum(alfajores$fabrica == 0) / totalA

cat("b) prob(Quilmes y 3 defectuosos):")
sum(alfajores$fabrica == 0 & alfajores$defectuosos == 3) / totalA

cat("c) prob puntual conjunta del vector XY:")
table(alfajores)/totalA

cat("d) esperanza y var de X:")
mean(alfajores$defectuosos)
var(alfajores$defectuosos)

cat("e) prob(3 defectuosos | Quilmes):")
sum(alfajores$fabrica == 0 & alfajores$defectuosos == 3) / sum(alfajores$fabrica == 0)

cat("f) prob(X=3):")
sum(alfajores$defectuosos == 3) / totalA

cat("g) prob(Quilmes | 3 defectuosos):")
sum(alfajores$fabrica == 0 & alfajores$defectuosos == 3) / sum(alfajores$defectuosos == 3)

cat("h) prob(Pilar | 3 defectuosos):")
sum(alfajores$fabrica == 1 & alfajores$defectuosos == 3) / sum(alfajores$defectuosos == 3)
```

# Ejercicio 2


```r
lamparas <- read.table("lamparas.txt")

cat("a) prob(duracion > 30hs):")
sum(lamparas>30) / length(lamparas[,])
```

```r
ej2b = ecdf(lamparas[,]) # F(x) = sum(valores < x) /n
plot(ej2b,col="red", main="b) funcion de distrib empirica lamparas", xlab="duracion (hs)", ylab="prob acum")
```

```r
q = quantile(ej2b, 1-0.9)

cat("c) Estos datos permiten estimar que el 90%% de las lámparas producidas por 
esta fábrica dura más de", q, "horas y el 10%% dura menos de", q, "horas") ##??? mismo valor????
```

```r
graduados <- scan("graduados.txt")
cat("a) media muestral:", mean(graduados), "y mediana muestral:", median(graduados), "\n")

cat("b) desvio estandar:", sd(graduados), "y distancia intercuartil:", IQR(graduados), "\n")
```

```r
h = hist(graduados, prob = TRUE, main="c) histograma graduados con curva normal")

xs = seq(min(graduados)-1, max(graduados)+1, length= 200)
normal = dnorm(xs, mean = mean(graduados), sd = sd(graduados))  #normal (~mu, ~sigma)

lines(xs, normal, lwd = 2)
```
