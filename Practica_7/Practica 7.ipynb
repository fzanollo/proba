# Ejercicio 2


###### Resumen - estimadores de la normal
a) $X_i\sim N(\theta,1)$ => $\hat{\theta}=\bar{X}=mean(datos)$

b) $X_i\sim N(0,\theta)$ => $\hat{\theta}=max(X_1...X_n)$

c) $X_i\sim N(\mu,\sigma^2)$ => $\hat{\mu}=\bar{X}$ y $\hat{\sigma}^2=\frac{\sum_{i=1}^n(X_i-\bar{X})^2}{n}$

```r
mu = runif(1,-10,10)
sigma2 = runif(1,1,10)
n = 40

obs = rnorm(n, mu, sqrt(sigma2))

cat('mu:', mu, 'sigma2:', sigma2, '\n')

cat('\na) mu(est)=', mean(obs))
cat('\nb) sigma2(est)=', var(obs)) #### ok?
cat('\nc) mu(est)=', mean(obs), 'sigma2(est)=', sum((obs-mean(obs))^2)/n )
```

# Ejercicio 4


###### Resumen - estimadores de la exponencial
- Estimador de momentos: $\hat{\theta} = \frac{1}{\bar{X}}$
- Estimador MV: igual al otro?

```r
lambda = runif(1, 1, 100)
n = round(runif(1, 10, 100))
datos4 = rexp(n, lambda)

cat("lambda=", lambda, "n=", n, "\nhead(datos)=", head(datos4))

#a
lambda_hat = 1/mean(datos4)
cat('\n\na) est. momentos=', lambda_hat, ' MV=', lambda_hat, " cuantil 0.9=", qexp(0.9, lambda_hat))
```

b) y c) en la carpeta

```r
#d
datos4d = scan(text="39.08 45.27 26.27 14.77 65.84 49.64 0.80 66.58 69.60 32.42 228.36 64.79 9.38 3.86 37.18 104.75 3.64 104.19 8.17 8.36")
lambda_hat = 1/mean(datos4d)
cat('d) \ni) est. momentos=', lambda_hat, ' MV=', lambda_hat, " cuantil 0.9=", qexp(0.9, lambda_hat))
cat("\nii) ordenando y buscando el punto donde queda el 0.9 de los datos de un lado??????")

#e
datos4e = scan("lamparas.txt")
lambda_hat = 1/mean(datos4e)
cat('\n\ne) \ni) est. momentos=', lambda_hat, ' MV=', lambda_hat, " cuantil 0.9=", qexp(0.9, lambda_hat))

distrib4e = ecdf(datos4e)
cat("\nii) percentil 0.9 estimado=", quantile(distrib4e, 0.9))
```

# Ejercicio 5


```r
lambda = runif(1, 20, 60)
n = 48
datos5 = rpois(n, lambda)

cat("lambda=", lambda, "n=", n, "\nhead(datos)=", head(datos5))

#a
lambda_hat = mean(datos5)
cat('\n\na) est. momentos= MV=', lambda_hat)

#b
cat('\n\nb) P(X>40)=', 1-ppois(40,lambda_hat))

#b
cat('\n\nc) estimado P(X>40)=', sum(datos5>40)/length(datos5))
```
