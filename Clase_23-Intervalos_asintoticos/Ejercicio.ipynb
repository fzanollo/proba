```r
calcular_intervalo = function(n, p_ = 0.38, alpha = 1-0.95){
    z = qnorm(1-alpha/2)

    raiz = sqrt( (p_ * (1-p_))/n )

    limInf = p_ - z * raiz
    limSup = p_ + z * raiz
    
    c(limInf, limSup)
}
```

```r
n=100
interv = calcular_intervalo(n)

cat("intervalo de confianza asintotico para p de la Bi(1,p), n=", n, "\n")
cat("[", interv, "]\n")

long = interv[2] - interv[1]
cat("long del intervalo =", long, '\n\n')

n=4020
while(long > 0.03){
    n=n+1
    interv = calcular_intervalo(n)
    long = interv[2] - interv[1]
    cat("para n=", n,"la long es=", long, '\n')
}

```
