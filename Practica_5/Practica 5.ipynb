```r
n_talque = function(valor, media_xi, var_xi, prob){
    a = media_xi
    b = sqrt(var_xi)*qnorm(1-prob)
    c = -valor
    
    n1 = (-b-sqrt(b^2-4*a*c))/(2*a)
    n2 = (-b+sqrt(b^2-4*a*c))/(2*a)
    
    max(n1,n2)^2
}

tcl_suma = function(media_xi, var_xi, n, valor){ # asume P(sum(X_i) < valor)
    pnorm((valor-media_xi*n) / sqrt(n*var_xi))
}

tcl_prom = function(media_xi, var_xi, n, valor){ # asume P(X_n < valor)
    pnorm((valor-media_xi) / sqrt(var_xi/n))
}
```
